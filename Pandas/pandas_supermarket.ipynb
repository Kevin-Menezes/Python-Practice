```python
df["Address"] = df["Address"] + "," + df["City"] + "," + df["State"] + "," + df["Country"] # Combining the full address in 1 column
df
```

```python
df["Coordinates"] = df["Address"].apply(arc.geocode) # Applying the geocode function to all rows in that column
df
```

```python
# Get latitude and longitude
df["Latitude"] = df["Coordinates"].apply(lambda x : x.latitude)
df["Longitude"] = df["Coordinates"].apply(lambda x : x.longitude)
df
```
